```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import integrate
```

```python
#Define electric field function
def fun(x, z, R):
    #Returns the function value 
    return (z-R*x)/(R**2+z**2-2*R*z*x)**(3/2)
```

```python
#Plotting electric field with scipy.integrate
#Define the bounds for integration
def scipy_int(z, R, bound1, bound2):
    electric_field = []
    #Function to integrate
    for h in z:
        f = lambda x: fun(x, h, R)
        #Saving the electric field
        electric_field.append(integrate.quad(f, bound1, bound2)[0])
    
    return electric_field
```

```python
#Define integrator using simpson's rule
def simpson(f, a, b):
    #Splitting into n subintervals
    if a == b:
        return 0
    if a < b:
        sign = 1
    else:
        sign = -1

    h = (a+b)/2 #Get the height
    fh = f(h) #Get the function value at the given height
    fa = f(a)
    fb = f(b)
    dx = (b-a)/6 #Set n=2, so 3*2
    #Integrate with simpson's rule
    i = dx*sign*(fa+4*fh+fb)
    
    return i, h, fh

#Keep iterating over guesses to allow convergence
def dynamic_simpson(f, a, b, h, fa, fb, fh, guess, tolerance, nmax, count=0):
    if count < nmax:
        guess_ah, left_h, left_fh = simpson(f, a, h)
        guess_bh, right_h, right_fh = simpson(f, h, b)
        err = np.abs(guess_ah+guess_bh-guess)
        
        if err < 15*tolerance:
            return guess, count
        else:
            #Provide smaller and smaller tolerances each time, and update the count
            #Separate the left and right side, up until h
            guess_left, count_left = dynamic_simpson(f, a, h, left_h, fa, fh, left_fh, guess_ah, tolerance/2, nmax, count+1)
            guess_right, count_right = dynamic_simpson(f, h, b, right_h, fh, fb, right_fh, guess_bh, tolerance/2, nmax, count+1)
            
            #Combine the two sides for total guess and count
            guess_t = guess_left+guess_right
            count_t = count_left+count_right
            
            return guess_t, count_t
        
    else:
        print('Does not converge. Increase tolerance or nmax')
        exit(1)
        
def summing(f, a, b, tolerance, nmax):
    i, h, fh = simpson(f, a, b)
    result= np.array(dynamic_simpson(f, a, b, h, f(a), f(b), fh, i, tolerance, nmax))
    #Add up the counts and integral for each interval of simpson
    count = np.sum(result[1::2])
    integ = np.sum(result[0::2])
    
    return integ, count

def integral(z, R, a, b, tolerance=1e-7, nmax=1000):
    electric_field = []
    #Function to integrate
    for h in z:
        f = lambda x: fun(x, h, R)
        try: 
            val, dummy = summing(f, a, b, tolerance, nmax)
        except: 
            val = 0
        #Saving the electric field
        electric_field.append(val)
        
    return electric_field
```

```python
#Generating the electric field of an infinitessimally thin spherical shell of charge with radius R
x = [-1, 1]
z = np.linspace(0, 2, 501)
R = 1
integral_simpson = integral(z, R, -1, 1,1e-7,1000)
e_field = scipy_int(z, R, x[0], x[1])
```

```python
#Plotting comparison between two integration methods
fig, ax = plt.subplots(1,1,figsize=(25,8))
ax.plot(z, integral_simpson, '-', markersize = '2', color = 'black', label = 'Simpson')
ax.plot(z, e_field, '--', markersize = '10', color = 'red', label = 'Scipy')
ax.set_title("Electric Field of Sphere", fontsize = 22)
ax.set_xlabel("z", fontsize = 18)
ax.set_ylabel("E-field", fontsize = 18)
plt.legend(fontsize=16)
plt.show()
```

There are two regions seen in this plot: $z<R$ and $z>R$, separated by where $z=1$. There is a singularity at $z=1$, where quad evaluates the input and gives an error, crashing the program. The simpson integrator fails at this point, which is why I included a try/except feature to avoid $z=1$.
