```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import interpolate
```

```python
#Chebyshev fitting
def cheb_mat_uniform(nx,ord):
    x = np.linspace(0.5,1,nx)
    mat = np.zeros([nx,ord+1])
    mat[:,0] = 1.0
    if ord>0:
        mat[:,1] = x
    if ord>1:
        for i in range(1,ord):
            #T_(n+1) = 2xT_(n)-T_(n-1)
            mat[:,i+1] = 2*x*mat[:,i]-mat[:,i-1] 
    return mat,x

n = 5000 #Num of data points (num of rows)
ord = 100 #Highest order to truncate the matrix at
mat,x = cheb_mat_uniform(n,ord) #Get the matrix
y = np.log2(x) #Function to fit
f = lambda x: np.log2(x)

#Least-squares fit (not working)
def least_sq(mat,y):
    lhs = np.dot(mat.transpose(),mat)
    rhs = np.dot(mat.transpose(),y)
    #Perform fit: (A^T A)^(-1)A^T y
    fitp = np.dot(np.linalg.pinv(lhs),rhs)
    return fitp

#Least-squares fit (working)
def lsq(A,x,y):
    d = np.matrix(y).transpose()
    u,s,vt = np.linalg.svd(A,False)
    sinv = np.matrix(np.diag(1.0/s))
    fitp = vt.transpose()*sinv*(u.transpose()*d)
    return fitp

#Define function to get the optimal coefficients given a specific tolerance
def coeff_ord(mat,c,tol,ind,yexpec):
    
    ind = len(c)
    for i in range(len(c)):
        pred = np.dot(mat[:,:i],fitp[:i])
        dif = pred - yexpec
        square = [(p-y)**2 for (p,y) in zip(pred,yexpec)]
        if (np.sqrt(np.mean(square))) <= 9*tol: 
            #Multiplying tol by 9 to catch when magnitude is 1e-6
            ind = i 
            break
    return ind
    
    #if (ind+1) == len(c):
    #    return ind
    #if np.sqrt(np.mean(square)) <= tol:
    #    return ind
    #else:
    #    return coeff_ord(mat,c,tol,ind+1,yexpec)
        
#Prediction with optimal coefficients
fitp = lsq(mat,x,y)
ncoeff = coeff_ord(mat,fitp,1e-6,0,y)
print('The number of terms needed to use a truncated Chebyshev is', ncoeff)
pred = np.dot(mat[:,:ncoeff],fitp[:ncoeff])
pred = np.squeeze(np.asarray(pred))
```

The number of terms that you need is one less than the given order. I tried to compare the tolerance with the coefficients, but it gave the same result.

```python
#Prediction with non-optimal coefficients
def cheb_fit(f,ord):
    x = np.linspace(0.5,1,ord+1)
    y = f(x)
    mat = np.zeros([ord+1,ord+1])
    mat[:,0] = 1
    mat[:,1] = x
    for i in range(1,ord):
        mat[:,i+1] = 2*x*mat[:,i]-mat[:,i-1]
    #p=yA^(-1)
    fitp = np.linalg.pinv(mat)@y
    return fitp

#Prediction with non-optimal coefficients
fitpp = cheb_fit(f,ord)  
ncoeff2 = coeff_ord(mat,fitpp,1e-6,0,y)
pred2 = np.dot(mat[:,:ncoeff2],fitpp[:ncoeff2])
pred2 = np.squeeze(np.asarray(pred2))
```

```python
#Plotting a Chebyshev fit with two different sets of coefficients
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2,gridspec_kw={'height_ratios':[1,0.3]},figsize=(25,8))

ax1.plot(x, y, '.', markersize = '6', color = 'red')
ax1.plot(x, pred, '-', markersize = '2', color = 'black')
ax1.set_title("Chebyshev Fit with Optimal Coefficients", fontsize = 20)
ax1.set_ylabel("f(x) = log2(x)", fontsize = 18)
ax1.set_xlabel("x", fontsize = 18)

ax2.plot(x, y, '.', markersize = '6', color = 'red')
ax2.plot(x, pred2, '-', markersize = '2', color = 'black')
ax2.set_title("Chebyshev Fit with Non-optimal Coefficients", fontsize = 20)
ax2.set_ylabel("f(x)  = log2(x)", fontsize = 18)
ax2.set_xlabel("x", fontsize = 18)

ax3.plot(x, pred-y, '-', markersize = '6', color = 'blue')
ax3.axhline(0,linestyle='--', linewidth = '2', color = 'black')
ax3.set_ylabel("Residuals", fontsize = 18)
ax3.set_xlabel("x", fontsize = 18)

ax4.plot(x, pred2-y, '-', markersize = '6', color = 'blue')
ax4.axhline(0,linestyle='--', linewidth = '2', color = 'black')
ax4.set_ylabel("Residuals", fontsize = 18)
ax4.set_xlabel("x", fontsize = 18)

plt.show()

print('rms error for chebyshev optimal coeff is',np.sqrt(np.mean((pred-y)**2)),'with max error',np.max(np.abs(pred-y)))
print('rms error for chebyshev non-optimal coeff is',np.sqrt(np.mean((pred2-y)**2)),'with max error',np.max(np.abs(pred2-y)))
```
