```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import interpolate
from scipy import integrate
import time
```

Solving for all decay products of pure U238. We saw the rk4 method in class, so let's try this method to approach the problem.

```python
#Solving the ODEs
#Collect the half lives of all of the decay processes in the chain

#Defining time conversion constants
y = 3.154e7
b = year*1e9
d = 86400
h = 3600
m = 60
us = 1e-6

#Starting with a pure sample of U238
U238 = [4.468*b, 24.1*d, 6.7*h, 245500*year, 75380*y, 1600*y, 3.8235*d, 
        3.1*m, 26.8*m, 19.9*m, 164.3*us, 22.3*y, 5.015*y, 138.376*d]

def fun(x,y,half_life=U238):
    dydx=np.zeros(len(half_life)+1)
    #For the first approximation do
    dydx[0]=-y[0]/(half_life[0])
    for i in range(1,len(dydx)-1):
        dydx[i]=y[i-1]/(half_life[i-1])-y[i]/(half_life[i])
    dydx[-1]=y[-2]/(half_life[-2])
    return dydx

#Initial values for each decay product; only have uranium238 to begin with
y0=np.asarray([1] + [0]*(len(U238))) 
x0=0 #Choosing an interesting range
x1=1e1*b
```

The rk4 method takes a long time to run; perhaps there are better methods to explore, such as the Radau. This method performs faster, so I will use it instead to avoid long wait times.

```python
#Solve the ODEs with Radau method
t1=time.time()
ans_stiff=integrate.solve_ivp(fun,[x0,x1],y0,method='Radau')
t2=time.time()
print('Took',ans_stiff.nfev,'evaluations and',t2-t1,'seconds to solve implicitly')
#Compare the results of the different integration methods
#Run time is too long so comment out
#print('Final values are',ans_rk4.y[0,-1],'and',ans_stiff.y[0,-1],'with truth',np.exp(-1*(x1-x0)))
```

```python
#Generating the values in the region from x0 to x1
plomb = ans_stiff.y[14,:] #Last step in the decay series
uranium_238 = ans_stiff.y[0,:]
thor = ans_stiff.y[4,:] #First step in the decay series
uranium_234 = ans_stiff.y[3,:]

#Ratio of Pb-206 to U-238 
ratio_plomb_uranium_238 = plomb/uranium_238
#Ratio of Th-230 to U-234
ratio_thor_uranium_234 = thor/uranium_234

time = ans_stiff.t[0:-1]
```

```python
#Plotting the ratio of plomb to uranium as a function of time
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(15,6))
ax1.plot(time, ratio_plomb_uranium_238[0:-1], '-', color='red', markersize=2)
ax1.set_title("Pb206/U238 as a function of time", fontsize = 20)
ax1.set_ylabel("Pb206/U238", fontsize = 18)
ax1.set_xlabel("Time (s)", fontsize = 18)

#Plotting the ratio of thorium to uranium as a function of time
ax2.plot(time, ratio_thor_uranium_234[0:-1], '-', color='blue', markersize=2)
ax2.set_title("Th234/U238 as a function of time", fontsize = 20)
ax2.set_ylabel("Th234/U238", fontsize = 18)
ax2.set_xlabel("Time (s)", fontsize = 18)
ax2.set_xlim(0, 5e13)

plt.show()
```

The ratio of Pb-206 to U-238 increases exponentially. This relation makes sense analytically since the quantity of decay product decreases as you traverse down the chain of exponentially decaying products. Because there is no loss term accounted for, the Pb-206 will accumulate U-238 decays indefinitely.

The ratio of Th-230 to U-234 increases and reaches an asymptote at aroud 0.3. This makes sense analytically as U-234 will accumulate exponentially while the amount of Th-230 is relatively small. After a while, Th-230 will also begin to accumulate exponentially. The growth rates will be roughly equal after a long period of time passes, which is why an equilibrium is observed in the plot above.
