```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy import signal
```

```python
#Defining a gaussian function
def gauss(x,mu,sig):
    return np.exp(-(x-mu)**2/(2*sig**2))

#Defining correlation function
def correlation(f,g):
    a = np.fft.fft(f)*np.conj(np.fft.fft(g))
    return np.fft.ifft(a)

#Define the two gaussians to take the correlation function of
#Define parameters
N = 100
mu = 0
sig = 1
x = np.linspace(-10,10,N)

#Generating the gaussians
g1 = gauss(x,mu,sig)
g2 = gauss(x,mu,sig)

#Plot the results
plt.plot(x, g1,color='black',label='original gaussian')
plt.plot(x, correlation(g1,g2),color='red',label='correlation function')
plt.xlabel('x values')
plt.ylabel('y values')
plt.savefig('q2_gauss_correlation')
plt.legend()
plt.show()
```
