```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy import signal
```

```python
#Defining a gaussian function
def gauss(x,mu,sig):
    return (np.exp(-(x-mu)**2/(2*sig**2)))

#Defining correlation function
def correlation(f,g):
    a = np.fft.fft(f)*np.conj(np.fft.fft(g))
    b = np.fft.ifft(a)
    return b.real/(max(b.real))

#Define the two gaussians to take the correlation function of
#Define parameters
N = 0.001
mu = 0
sig = 1
x = np.arange(-5,5,N)

#Generating the gaussians
g = gauss(x,mu,sig)
g_corr = correlation(g,g)
#gg = np.correlate(g,g,'same')/max(np.correlate(g,g,'same'))

#Plot the results
#plt.plot(x,gg,color='green')
plt.plot(x,g,color='black',label='original gaussian')
plt.plot(x,g_corr,'--',color='red',label='correlation function')
plt.xlabel('x values')
plt.ylabel('y values')
plt.savefig('q2_gauss_correlation')
plt.legend(loc='upper right')
plt.show()
```

There are two gaussians! The max overlap is at the ends; in red we see two gaussians to indicate this.
