```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy import signal
```

```python
DX = 0.001
MU = 0
SIG = 1
MINV = -5
MAXV = 10

def conv(f,g):
    ft = np.fft.fft(np.pad(f, [0, 10]))
    gt = np.fft.fft(np.pad(g, [0, 10]))
    
    y_new = np.real(np.fft.ifft(ft*gt))
    return (y_new/max(y_new)).real
    
#Defining a gaussian function to shift
def gauss(x,mu,sig):
    return (np.exp(-(x-mu)**2/(2*sig**2)))

#Making the gaussian array
x_val = np.arange(MINV,MAXV,DX)

#Defining functions
g = gauss(x_val,MU,SIG)
y = (1/5)*x_val
t = np.ones([len(x_val)])*0.5
cos = np.cos(x_val)
sin = np.sin(x_val)
```

```python
c = conv(g,cos)
plt.plot(x_val,g,label='gaussian',color='black')
plt.plot(x_val,cos,label='cosine',color='blue')
plt.plot(x_val,c[:len(x_val)],label='convolution',color='red')
plt.xlabel('x values')
plt.ylabel('y values')
plt.ylim(-1.6,1.6)
plt.legend()
plt.savefig('q4_no_wrapping_cos')
plt.show()
```

```python
c = conv(g,t)
plt.plot(x_val,g,label='gaussian',color='black')
plt.plot(x_val,t,label='line',color='blue')
plt.plot(x_val,c[:len(x_val)],label='convolution',color='red')
plt.xlabel('x values')
plt.ylabel('y values')
plt.ylim(-1.6,1.6)
plt.legend()
plt.savefig('q4_no_wrapping_line')
plt.show()
```

```python
c = conv(sin,y)
plt.plot(x_val,y,label='line',color='black')
plt.plot(x_val,sin,label='sine',color='blue')
plt.plot(x_val,c[:len(x_val)],label='convolution',color='red')
plt.xlabel('x values')
plt.ylabel('y values')
plt.ylim(-2,2)
plt.legend()
plt.savefig('q4_no_wrapping_sin_line')
plt.show()
```
