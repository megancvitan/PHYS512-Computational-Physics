```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy import signal
from sympy import DiracDelta, Subs, Symbol
from sympy.abc import x,y

DX = 0.01
MU = 0
SIG = 1
MINV = -5
MAXV = 5

#Defining function to shift an array by an arbitrary amount using a convolution
def shift(arr,s):
    N = len(arr) #Length of the gaussian passed in
    kvec = np.arange(N)
    yft = np.fft.fft(arr)
    J = np.complex(0,1)
    
    #Considering the shift value s in the phase
    yft_new = yft*np.exp(-2*np.pi*J*s*kvec/N)
    #Getting the inverse ft with the s included now
    y_new = (np.fft.ifft(yft_new)).real
    return y_new
    
#Defining a gaussian function to shift
def gauss(x,mu,sig):
    return (np.exp(-(x-mu)**2/(2*sig**2)))/(sig*np.sqrt(2*np.pi))

#Making the gaussian array
x_val = np.arange(MINV,MAXV,DX)
g = gauss(x_val,MU,SIG)
#Defining the shift
s = len(x_val)/2 
print('The length of half of the gaussian array is',s)
#Getting the shifted gaussian values
g_s = shift(g,s)
```

```python
#Plot the gaussian shift
plt.plot(x_val,g,color='black',label='original gauss')
plt.plot(x_val,g_s,'--',color='red',label='shifted gauss')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.savefig('q1_gauss_shift')
plt.show()
```

We can see that the new gaussian is shifted by the correct amount. Let's change the x axis bounds to see the whole gaussian shift.

```python
x_val = np.arange(MINV,50,DX)
g = gauss(x_val,MU,SIG)
#Defining the shift
s = len(x_val)/2 
g_s = shift(g,s)

#Plot the gaussian shift
plt.plot(x_val,g,color='black',label='original gauss')
plt.plot(x_val,g_s,'--',color='red',label='shifted gauss')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.savefig('q1_gauss_shift_2')
plt.show()
```
